```python
terrain_HARV = rx.open_rasterio("data/NEON-DS-Airborne-Remote-Sensing/HARV/DTM/HARV_dtmCrop_WGS84.tif")
```

```python
terrain_HARV
```

```python
surface_HARV.rio.crs
```

```python
terrain_HARV.rio.crs
```

```python
print(surface_HARV.rio.bounds())
```

```python
print(terrain_HARV.rio.bounds())
```

```python
terrian_HARV = terrain_HARV.rio.reproject_match(surface_HARV) # reproject lat-long to utm and snap the boundary to align
```

```python
terrain_HARV.rio.nodata
```

```python
terrian_HARV.plot(robust=True)
```

```python
terrian_HARV.rio.reproject_match? # resampling method, default use nearest neighbor
```

```python
surface_HARV.plot()
```

```python
canopy_height_HARV = surface_HARV - terrian_HARV # to get canopy height: extract elevation from surface height 
```

```python
canopy_height_HARV.plot(robust=True, figsize=(12, 8)) # robust=True rmove NA values, can see a pond, trail/road
```
